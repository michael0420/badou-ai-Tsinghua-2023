## 转换位灰度图片

```python
def cv_show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
cv_show('test',img)

# 转换为灰度图像
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv_show('gray',gray_img)
```

```python
gray_img.shape
```

```python
# 手动转换为灰度图像
h,w = img.shape[:2] #行列                              
gray_img_1 = np.zeros([h,w],np.uint8)                
for i in range(h):
    for j in range(w):
        m = img[i,j]                            
        gray_img_1[i,j] = int(m[0]*0.11 + m[1]*0.59 + m[2]*0.3) 
cv_show('gray',gray_img_1)
```
